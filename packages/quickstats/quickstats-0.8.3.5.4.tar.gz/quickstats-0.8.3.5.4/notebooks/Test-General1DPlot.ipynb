```python
import numpy as np
import pandas as pd
x1 = np.linspace(4, 10, 100)
y1 = (x1 - 7) ** 2
yerr1 = np.sqrt(y1)

x2 = np.linspace(4, 10, 100)
y2 = (x1 - 6) ** 2
yerr2 = np.sqrt(y2)
df = pd.DataFrame({'x': x1, 'y': y1})
dfs = {
    'A': pd.DataFrame({'x': x1, 'y': y1, 'yerrhi': yerr1, 'yerrlo': yerr1}),
    'B': pd.DataFrame({'x': x2, 'y': y2, 'yerrhi': yerr2, 'yerrlo': yerr2})
}
```

```python
from quickstats.plots import General1DPlot
```

```python
styles_map = {
    'A': {
        'plot': {
            'marker': 'o',
            'markerfacecolor': 'none'
        }
    },
    'B': {
        'plot': {
            'linestyle': '--'
        }        
    }
}
styles = {
    'plot': {
        'marker': 'none'
    }
}
label_map = {
    'A.error': 'Uncertainty'
}
config = {
    'isolate_error_legend': False,
    'inherit_color': True,
    'error_on_top': False
}
config_map = {
    'A': {
        'isolate_error_legend': True
    }
}
plotter = General1DPlot(dfs, styles=styles, styles_map=styles_map, config=config,
                        config_map=config_map, label_map=label_map)
```

```python
plotter.draw('x', 'y', 'yerrhi', 'yerrlo', ypad=0.2)
```
