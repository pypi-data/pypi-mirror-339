```python
from quickstats.plots import AbstractPlot
```

```python
plotter = AbstractPlot()
```

```python
plotter.reset_artists()
plotter.add_point(0.44, 0.72, label='test')
plotter.add_point([0.2, 0.22, 0.25, 0.28, 0.31], [0.8]*5, color='r', label='test2')
plotter.add_vline(0.35, color='b', linestyle='--', label='myline')
plotter.add_text('hello//world', x=0.2, y=0.3, dy=0.02)
plotter.add_annotation('custom', xy=(0.32, 0.66), label='annotation')
plotter.add_fill_between([0.25, 0.37, 0.46], [0.615, 0.675, 0.633], [0.668, 0.725, 0.713], label='fill')
plotter.add_errorband([0.21, 0.32, 0.40], [0.635,  0.712, 0.689], [0.615, 0.675, 0.633], [0.668, 0.725, 0.713], label='band')
plotter.add_errorband(x=[0.40, 0.40], y=[0.65, 0.7], xerr=[0.38, 0.42], yerrlo=[0.65, 0.65], yerrhi=[0.7, 0.7], label='band2')
```

```python
plotter.artists
```

```python
plotter.draw(xlabel='xlabel', ylabel='ylabel', ypad=0.3)
```
