```python
import matplotlib.pyplot as plt
import numpy as np

from edges_cal import xrfi

%matplotlib inline
```

```python
sky = np.ones(1000)
```

```python
noise = np.random.normal(scale=0.05, size=1000)
```

```python
rfi = np.zeros(1000)
np.random.seed(12345)
rfi[np.random.randint(0, len(rfi), 20)] = 200
```

```python
true_flags = rfi > 0
```

```python
total_sky = sky + noise + rfi
```

```python
freq = np.linspace(50, 150, 1000)
```

```python
flags, info = xrfi.xrfi_model(total_sky, freq=freq, std_estimator="sliding_rms")
```

```python
plt.plot(total_sky)
plt.plot(info.models[0](x=freq))
```

```python
for std in info.stds:
    plt.plot(std)
```

```python
flags != true_flags
```
