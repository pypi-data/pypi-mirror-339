```python
import simetri.graphics as sg

canvas = sg.Canvas()

logo = sg.Batch([sg.logo()])
canvas.scale(0.75)
sidle = sg.frieze.sidle(logo.copy(), ((55, 0), (55, 1)), 200, 3)
canvas.draw(sidle.translate(0, 250))

hop = sg.frieze.hop(logo.copy(), (100, 0), 6)
canvas.draw(hop)

jump = sg.frieze.jump(logo.copy(), ((0, 65), (1, 65)), 100, 6)
canvas.draw(jump.translate(0, -400))

step = sg.frieze.step(logo.copy(), ((0, 65), (1, 65)), 50, 6)
canvas.draw(step.translate(0, -800))

spinning_hop = sg.frieze.spinning_hop(logo.copy(), (15, 40), 120, 0, 6)
canvas.draw(spinning_hop.translate(0, -1200))

spinning_jump = sg.frieze.spinning_jump(logo.copy(), ((0, 65), (1, 65)),
                                  ((55, 0), (55, 1)), 200, 3)
canvas.draw(spinning_jump.translate(0, -1600))

spinning_sidle = sg.frieze.spinning_sidle(logo.copy(), ((55, 0), (55, 1)),
                                   ((0, 65), (1, 65)), 100, 200, 3)
canvas.draw(spinning_sidle.translate(0, -2000))

canvas.display()
```
