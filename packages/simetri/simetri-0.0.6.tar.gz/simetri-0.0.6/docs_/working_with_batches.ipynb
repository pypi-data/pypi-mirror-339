# Batch Objects

Batch objects are used for grouping Shape objects or other Batch objects or both. They can also be used for setting styling attributes of multiple objects at once. Applying multiple transformations to a Shape object results in a Batch object containing transformed copies of the target Shape object. But applying multiple transformations to a Batch object does not yield a new Batch object.

### Transforming Shape objects with repetitions yield Batch objects that contain the target Shape as well as the transformed copies of it.

```python
import simetri.graphics as sg

canvas = sg.Canvas()

points  = [(0, 0), (50, 50)]
segment = sg.Shape(points)
print(segment)
print(id(segment))
canvas.draw(segment)
segments = segment.mirror(segment.right, reps=1)
# segments is the batch returned by the mirror method

print(segments)
print(id(segments))
canvas.draw(segments)
segments.translate(0, -100)
segments.translate(100, 0, reps=4)

# since segments is a batch, transforming it with multiple
# repetitions does not yield a new batch
print(id(segments)) # same object as before translation
print(id(segments[0])) # this is the same object as before translation

canvas.draw(segments)
canvas.display()
```

### Batch objects can be used to group multiple objects. 

```python
import simetri.graphics as sg

canvas = sg.Canvas()

hex = sg.reg_poly_shape((0, 0), 6, 30)
oct = sg.reg_poly_shape((0, 70), 8, 30)

polygons = sg.Batch([hex, oct])
canvas.draw(polygons)
canvas.translate(100, 0)

# Now we can transform all shapes in the batch at once
scale = sg.scale_matrix(.9, .9)
translate = sg.translation_matrix(80, 0)
transform = translate @ scale
polygons.transform(transform, reps=4)

# We can change attributes of all shapes in the batch at once
polygons.set_attribs('fill_color', sg.red)
canvas.draw(polygons)

canvas.display()
```

### Batches behave similar to Python lists, but they are not lists.

```python
import simetri.graphics as sg

canvas = sg.Canvas()

hex = sg.reg_poly_shape((0, 0), 6, 30)
oct = sg.reg_poly_shape((0, 70), 8, 30)
sqr = sg.reg_poly_shape((0, 140), 4, 30)
pent = sg.reg_poly_shape((0, 210), 5, 30)
deca = sg.reg_poly_shape((200, 210), 10, 30)

batch1 = sg.Batch([hex, oct])
batch2 = sg.Batch([sqr, pent])

all_polygons = batch1 + batch2
all_polygons[0].translate(100, 0) # same as all_poygons.elements[0].translate(100, 0)

batch3 = sg.Batch(all_polygons[1:3])
# now the sqr and hex are in two different batches
batch3.translate(200, 0)
all_polygons.append(deca)

canvas.draw(all_polygons)

canvas.display()
```

### Batches can be created empty and can contain other batches.

```python
import simetri.graphics as sg

canvas = sg.Canvas()

hex = sg.reg_poly_shape((0, 0), 6, 30)
oct = sg.reg_poly_shape((0, 70), 8, 30)
sqr = sg.reg_poly_shape((0, 140), 4, 30)
pent = sg.reg_poly_shape((0, 210), 5, 30)
deca = sg.reg_poly_shape((200, 210), 10, 30)

batch1 = sg.Batch() # empty batch
batch1.extend([hex, oct])
batch2 = sg.Batch([sqr, pent])

all_polygons = sg.Batch([batch1, batch2]) # a batch of batches
all_polygons[0].translate(100, 0) # translate the first batch

all_polygons.append(deca) # add a new shape to the batch of batches
# now all_polygons contain two batches and one shape

canvas.draw(all_polygons)

canvas.display()
```

### Transforming a Batch with multiple repetitions grow the Batch itself, unlike transforming Shape objects.

```python
import simetri.graphics as sg

canvas = sg.Canvas()

hex = sg.reg_poly_shape((0, 0), 6, 30)
oct = sg.reg_poly_shape((0, 70), 8, 30)

polygons = sg.Batch([hex, oct])
print(id(polygons))
polygons[-1].fill_color = sg.green

polygons.mirror(polygons.right, reps=1)
polygons[0].fill_color = sg.red

polygons.mirror(polygons.bottom, reps=1)
print(id(polygons))
# polygons is the same object as before mirroring

canvas.draw(polygons)
# Copies of the shapes retain the same attributes as the original

canvas.display()
```

### `batch.all_shapes` return a flat list of all shapes in nested batches.

```python
import simetri.graphics as sg

canvas = sg.Canvas()

hex = sg.reg_poly_shape((0, 0), 6, 30)
oct = sg.reg_poly_shape((0, 70), 8, 30)
sqr = sg.reg_poly_shape((0, 140), 4, 30)
pent = sg.reg_poly_shape((0, 210), 5, 30)
deca = sg.reg_poly_shape((200, 210), 10, 30)
dodec = sg.reg_poly_shape((200, 0), 12, 30, fill_color=sg.yellow)

batch1 = sg.Batch() # empty batch
batch1.extend([hex, oct])
batch2 = sg.Batch([sqr, pent])
batch3 = sg.Batch([deca])

batch2.append(batch3) # batch of batches

all_polygons = sg.Batch([batch1, batch2]) # nested batches
all_polygons.insert(0, dodec)

print(f'Number of elements in all_polygons: {len(all_polygons)}') # same as len(all_polygons.elements)

all_shapes = all_polygons.all_shapes # Shapes are stored in a flat list
print(f'Number of shapes in all_polygons: {len(all_shapes)}') # same as len(all_polygons.shapes)


def list_elements(elements, indent=0):
    for element in elements:
        items = 'elements'
        if isinstance(element, sg.Shape):
            items = 'vertices'
        print(f"{' ' * indent}{element.type} with {len(element)} {items}.")
        if isinstance(element, sg.Batch):
            list_elements(element.elements, indent + 2)

print('\nList of all_polygons.elements:')
print(f'all_polygons has {len(all_polygons)} elements:  {[e.type.value for e in all_polygons.elements]}')
list_elements(all_polygons.elements, 2)

all_polygons[1].set_attribs('fill_color', sg.red)
all_polygons[2].set_attribs('fill_color', sg.green)
canvas.draw(all_polygons)

# draw boundary-boxes of the batch1 and batch2
canvas.draw(sg.Shape(all_polygons[1].corners, closed=True, fill=False))
canvas.draw(sg.Shape(all_polygons[2].corners, closed=True, fill=False))

canvas.display()
```
